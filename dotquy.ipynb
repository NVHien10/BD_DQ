```python
# Đọc dữ liệu từ tập tin CSV
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Dotquy/data_dotquy.csv')
df
```

```python

# Kiểm tra xem tuổi có phải là số nguyên hay không
df['Is_Whole_Number'] = df['age'].apply(lambda x: x.is_integer())

# Lọc các hàng mà tuổi không phải là số thực (chỉ giữ lại số nguyên)
df = df[df['Is_Whole_Number']]

# Xóa cột tạm thời 'Is_Whole_Number'
df = df.drop(columns=['Is_Whole_Number'])

print(df)
```

```python

# Định nghĩa các khoảng tuổi và các nhãn cho các nhóm tuổi
bins = [0, 18, 35, 60, 100]
labels = ['0', '1', '2', '3']

# Chia cột "Age" thành các nhóm tuổi
df['Age Group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)

df
```

```python
df.info()
```

```python
# Chuyển đổi các nhóm tuổi thành các giá trị số
df['Age Group'] = df['Age Group'].cat.codes.astype(float)
df
```

```python
# Hiển thị dữ liệu trước khi loại bỏ các hàng có giá trị N/A
print("Trước khi loại bỏ các hàng có giá trị N/A trong cột BMI:")
print(df)

# Bước 3: Loại bỏ các hàng có giá trị N/A trong cột BMI
df = df.dropna(subset=['bmi'])

# Hiển thị dữ liệu sau khi loại bỏ các hàng có giá trị N/A
print("\nSau khi loại bỏ các hàng có giá trị N/A trong cột BMI:")
print(df)

```
