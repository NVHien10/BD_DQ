```python
import hashlib
df = pd.read_csv('/content/drive/MyDrive/data_dotquy.csv')
```

```python
for index, row in df.iterrows():
    work_type = row['work_type']
    hash_object = hashlib.sha256(work_type.encode())
    hash_value = hash_object.hexdigest()
    df.at[index, 'work_type'] = hash_value
df
```
